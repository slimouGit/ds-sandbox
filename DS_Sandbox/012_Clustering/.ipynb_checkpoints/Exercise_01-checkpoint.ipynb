```python
import pandas as pd
df = pd.read_csv("edlich-kmeans-A0.csv")
```

```python
df.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter(df.iloc[:,0], df.iloc[:,1], df.iloc[:,2], c='b', marker='o')
ax.view_init(45, 45)
plt.show()
```

# A) What is the best k?

```python
from sklearn.cluster import KMeans

kdf = pd.DataFrame({"k": range(1,11)})
kdf["kmeans"] = kdf.apply( lambda row: KMeans(n_clusters=row["k"], random_state=0).fit(df.values), axis=1)
kdf["Inertia"] = kdf.apply( lambda row: row["kmeans"].inertia_, axis=1)
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The Elbow Method showing the optimal k')
ax.set_xlabel('k')
ax.set_ylabel('Inertia')

ax.plot(kdf["k"], kdf["Inertia"], 'b')
plt.show()
```

# B) Can you print the clustering vector?

```python
#from elbow-method solution above I choose 6 as the best value of K
k = 6
rangecluster = range(0,k)
kmeans = kdf["kmeans"][k-1]

# get dataframe of cluster centers
bkdf = pd.DataFrame( kmeans.cluster_centers_, columns=["x", "y", "z"] )

# setup plot
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# plot cluster center points
ax.scatter(bkdf.iloc[:,0], bkdf.iloc[:,1], bkdf.iloc[:,2], c='black', marker='*', s=200)

# default colormap
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

# plot by cluster in separate colors
for cluster in rangecluster:
    # select cluster
    cdf = df.loc[df["cluster"] == cluster]
    
    # plot cluster
    ax.scatter(cdf.iloc[:,0], cdf.iloc[:,1], cdf.iloc[:,2], c=colors[cluster], marker='o')
    
    # cluster label
    ax.text(bkdf.iloc[cluster,0], bkdf.iloc[cluster,1], bkdf.iloc[cluster,2]+4, "Cluster {}".format(cluster), None, va='center', ha='center', color=colors[cluster], weight='bold', size='15')

ax.view_init(45, 45) 

# output
plt.show()
```
